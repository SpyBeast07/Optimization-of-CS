# Fibonacci Series Code

```python
def Fibonacci(n):

	# Check if input is 0 then it will
	# print incorrect input
	if n < 0:
		print("Incorrect input")

	# Check if n is 0
	# then it will return 0
	elif n == 0:
		return 0

	# Check if n is 1,2
	# it will return 1
	elif n == 1 or n == 2:
		return 1

	else:
		return Fibonacci(n-1) + Fibonacci(n-2)
```

```python
def fCalc(func, x):
    expression = func
    # print(expression)

    # x = 4.96
    # x = float(input("Enter the value of x:"))
    # print("x =", x)

    result = eval(expression)
    # print("f(x) = ", result)

    # result = "{:.2f}".format(result)

    return result
```

```python
def fibonacci(func, tb, a, b, n):
    tb.at[1, 'A'] = a
    tb.at[1, 'B'] = b

    # tb.at[1, 'r'] = Fibonacci(n-2)/Fibonacci(n)
    # tb.at[1, 'x1'] = tb.at[1, 'A'] + ((tb.at[1, 'r'])*((tb.at[1, 'B'])-(tb.at[1, 'A'])))  # x1
    # tb.at[1, 'x2'] = tb.at[1, 'B'] - ((tb.at[1, 'r'])*((tb.at[1, 'B'])-(tb.at[1, 'A'])))  # x2

    i = n
    while (i > 3):
        tb.at[(n-i+1), 'r'] = Fibonacci(i-2)/Fibonacci(i)
        tb.at[(n-i+1), 'x1'] = tb.at[(n-i+1), 'A'] + ((tb.at[(n-i+1), 'r'])*((tb.at[(n-i+1), 'B'])-(tb.at[(n-i+1), 'A'])))  # x1
        tb.at[(n-i+1), 'x2'] = tb.at[(n-i+1), 'B'] - ((tb.at[(n-i+1), 'r'])*((tb.at[(n-i+1), 'B'])-(tb.at[(n-i+1), 'A'])))  # x2
        tb.at[(n-i+1), 'f(x1)'] = fCalc(func, tb.at[(n-i+1), 'x1'])
        tb.at[(n-i+1), 'f(x2)'] = fCalc(func, tb.at[(n-i+1), 'x2'])

        if (tb.at[(n-i+1), 'f(x1)'] > tb.at[(n-i+1), 'f(x2)']):
            tb.at[(n-i+1)+1, 'A'] = tb.at[(n-i+1), 'x1']
            tb.at[(n-i+1)+1, 'B'] = tb.at[(n-i+1), 'B']
        else:
            tb.at[(n-i+1)+1, 'A'] = tb.at[(n-i+1), 'A']
            tb.at[(n-i+1)+1, 'B'] = tb.at[(n-i+1), 'x2']

        i -= 1
        
    tb.at[(n-i+1), 'r'] = Fibonacci(i-2)/Fibonacci(i)
    tb.at[(n-i+1), 'x1'] = tb.at[(n-i+1), 'A'] + ((tb.at[(n-i+1), 'r'])*((tb.at[(n-i+1), 'B'])-(tb.at[(n-i+1), 'A'])))  # x1
    tb.at[(n-i+1), 'x2'] = tb.at[(n-i+1), 'B'] - ((tb.at[(n-i+1), 'r'])*((tb.at[(n-i+1), 'B'])-(tb.at[(n-i+1), 'A'])))  # x2
    tb.at[(n-i+1), 'f(x1)'] = fCalc(func, tb.at[(n-i+1), 'x1'])
    tb.at[(n-i+1), 'f(x2)'] = fCalc(func, tb.at[(n-i+1), 'x2'])
    print("Optimal x =", (tb.at[(n-i+1), 'x1']+tb.at[(n-i+1), 'x2'])/2)
    print("Optimal f(x) =", (tb.at[(n-i+1), 'f(x1)']+tb.at[(n-i+1), 'f(x2)'])/2)

    return tb
```

# Question: x**2, [-5, 15], Iterations = 7

```python
func = 'x**2'
a = -5
b = 15
n = 7
tb = pd.DataFrame(columns=('r', 'A', 'B', 'x1', 'x2', 'f(x1)', 'f(x2)'))

fibonacci(func, tb, a, b, n)
```
