# Golden Section Code

```python
def fCalc(func, x):
    expression = func
    # print(expression)

    # x = 4.96
    # x = float(input("Enter the value of x:"))
    # print("x =", x)

    result = eval(expression)
    # print("f(x) = ", result)

    # result = "{:.2f}".format(result)

    return result
```

```python
def golden_ratio(func, tb, a, b):
    tb.at[1, 'A'] = a
    tb.at[1, 'B'] = b

    i = 1
    while (abs(tb.at[i, 'A'] - tb.at[i, 'B']) > 0.01):
        tb.at[i, 'x1'] = tb.at[i, 'A'] + (0.618*((tb.at[i, 'B'])-(tb.at[i, 'A'])))  # x1
        tb.at[i, 'x2'] = tb.at[i, 'B'] - (0.618*((tb.at[i, 'B'])-(tb.at[i, 'A'])))  # x2
        tb.at[i, 'f(x1)'] = fCalc(func, tb.at[i, 'x1'])
        tb.at[i, 'f(x2)'] = fCalc(func, tb.at[i, 'x2'])


        if (tb.at[i, 'f(x1)'] < tb.at[i, 'f(x2)']):
            tb.at[i+1, 'A'] = tb.at[i, 'x2']
            tb.at[i+1, 'B'] = tb.at[i, 'B']
        else:
            tb.at[i+1, 'A'] = tb.at[i, 'A']
            tb.at[i+1, 'B'] = tb.at[i, 'x1']

        i += 1

    tb.at[i, 'x1'] = tb.at[i, 'A'] + (0.618*((tb.at[i, 'B'])-(tb.at[i, 'A'])))  # x1
    tb.at[i, 'x2'] = tb.at[i, 'B'] - (0.618*((tb.at[i, 'B'])-(tb.at[i, 'A'])))  # x2
    tb.at[i, 'f(x1)'] = fCalc(func, tb.at[i, 'x1'])
    tb.at[i, 'f(x2)'] = fCalc(func, tb.at[i, 'x2'])
    print("Optimal x =", (tb.at[i, 'x1']+tb.at[i, 'x2'])/2)
    print("Optimal f(x) =", (tb.at[i, 'f(x1)']+tb.at[i, 'f(x2)'])/2)

    return tb
```

```python
def golden_ratio2(func, tb, a, b):
    tb.at[1, 'A'] = a
    tb.at[1, 'B'] = b

    i = 2

    tb.at[1, 'x1'] = tb.at[1, 'A'] + (0.618*((tb.at[1, 'B'])-(tb.at[1, 'A'])))  # x1
    tb.at[1, 'x2'] = tb.at[1, 'B'] - (0.618*((tb.at[1, 'B'])-(tb.at[1, 'A'])))  # x2
    tb.at[1, 'f(x1)'] = fCalc(func, tb.at[1, 'x1'])
    tb.at[1, 'f(x2)'] = fCalc(func, tb.at[1, 'x2'])

    while (abs(tb.at[i-1, 'A'] - tb.at[i-1, 'B']) <= 0.01 or i != 11):
        if (tb.at[i-1, 'f(x1)'] < tb.at[i-1, 'f(x2)']):
            tb.at[i, 'A'] = tb.at[i-1, 'x2']
            tb.at[i, 'B'] = tb.at[i-1, 'B']

            tb.at[i, 'x1'] = tb.at[i, 'A'] + (0.618*((tb.at[i, 'B'])-(tb.at[i, 'A'])))  # x1
            tb.at[i, 'x2'] = tb.at[i-1, 'x1']
            tb.at[i, 'f(x1)'] = fCalc(func, tb.at[i, 'x1'])
            tb.at[i, 'f(x2)'] = fCalc(func, tb.at[i, 'x2'])
        else:
            tb.at[i, 'A'] = tb.at[i-1, 'A']
            tb.at[i, 'B'] = tb.at[i-1, 'x1']
            tb.at[i, 'x1'] = tb.at[i-1, 'x2']
            tb.at[i, 'x2'] = tb.at[i, 'B'] - (0.618*((tb.at[i, 'B'])-(tb.at[i, 'A'])))  # x2
            tb.at[i, 'f(x1)'] = tb.at[i-1, 'f(x2)']
            tb.at[i, 'f(x2)'] = fCalc(func, tb.at[i, 'x2'])

        i += 1

    print("Optimal x =", (tb.at[i-1, 'x1']+tb.at[i-1, 'x2'])/2)
    print("Optimal f(x) =", (tb.at[i-1, 'f(x1)']+tb.at[i-1, 'f(x2)'])/2)

    return tb
```
